#Check Missing Values & Duplicates

```python
# Step 1: Check missing values (NaN) in each column
missing_counts = df.isnull().sum().sort_values(ascending=False)
missing_percent = (df.isnull().mean() * 100).round(2).sort_values(ascending=False)

# Combine both into one table for clarity
missing_report = pd.concat([missing_counts, missing_percent], axis=1)
missing_report.columns = ['Missing_Values', 'Missing_%']

# Display the missing value report
print(" Missing Values Report:\n")
display(missing_report)

# Step 2: Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"\n Total Duplicate Rows: {duplicate_count}")
```

#Remove Duplicates & Parse Date Column

```python
# Step 1: Remove exact duplicate rows
before = df.shape
df = df.drop_duplicates().reset_index(drop=True)
after = df.shape
print(f"Removed duplicates: {before[0] - after[0]} rows removed.")
print(f"New shape of dataset: {after}")

# Step 2: Detect and convert the Date column
# Find column name that contains the word 'date'
date_col = None
for col in df.columns:
    if 'date' in col.lower():
        date_col = col
        break

if date_col:
    print(f"Found date column: '{date_col}' â€” converting to datetime format...")
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    print(f"Conversion done. Missing date values: {df[date_col].isnull().sum()}")
    # Rename to 'Date' for consistency
    df.rename(columns={date_col: 'Date'}, inplace=True)
else:
    print("No column found with 'date' in its name. Check your dataset manually.")

# Step 3: Verify result
df[['Date']].head()
```

#Numeric checks, fill small missing values, choose ridership column

```python
# Step 1: List numeric columns and show basic stats
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numeric_cols)
display(df[numeric_cols].describe().T)

# Step 2: Fill small missingness (<=5%) with median, but just report larger gaps
for c in numeric_cols:
    pct_missing = df[c].isnull().mean() * 100
    if pct_missing == 0:
        continue
    elif pct_missing <= 5:
        med = df[c].median()
        df[c] = df[c].fillna(med)
        print(f"Filled {pct_missing:.2f}% missing in '{c}' with median = {med}")
    else:
        print(f" Column '{c}' has {pct_missing:.2f}% missing â€” handle this manually (drop/fill/impute)")

# Step 3: Choose ridership column (prefer 'Passengers')
if 'Passengers' in df.columns:
    ridership_col = 'Passengers'
else:
    # fallback: look for columns with 'pass' in the name
    candidates = [c for c in numeric_cols if 'pass' in c.lower()]
    if candidates:
        ridership_col = candidates[0]
    else:
        # final fallback: numeric with highest variance (likely counts)
        ridership_col = df[numeric_cols].var().sort_values(ascending=False).index[0]
print("\nUsing ridership column for analysis:", ridership_col)

# Optional: If you have a 'Line' or 'Metro_Line' column already, confirm it
if 'Line' in df.columns or 'Metro_Line' in df.columns:
    line_col = 'Line' if 'Line' in df.columns else 'Metro_Line'
    print("Found existing line column:", line_col)
else:
    print("No 'Line' column found. If you want to analyze by metro line color, add a 'Line' column mapping each station to its line (see later cell).")

# Preview cleaned numeric subset
display(df[[ridership_col] + (['Line'] if 'Line' in df.columns else [])].head())
```

#Create time features & distance bins

```python
# Ensure Date column exists
assert 'Date' in df.columns, "Date column missing â€” run the date parsing cell."

# Create common time features
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['weekday'] = df['Date'].dt.day_name()
df['is_weekend'] = df['weekday'].isin(['Saturday', 'Sunday'])

# If Date includes time, extract hour (if not present already)
if df['Date'].dt.time.notna().any():
    try:
        df['Hour'] = df['Date'].dt.hour
    except Exception as e:
        # fallback: only create Hour if any non-null time component exists
        print("Hour extraction skipped (no time component or parsing issue).", e)

# Distance bins (only if Distance_km exists)
if 'Distance_km' in df.columns:
    max_d = df['Distance_km'].max()
    bins = [-0.1, 1, 3, 6, 20, max_d + 1]
    labels = ['very_short', 'short', 'medium', 'long', 'very_long']
    df['distance_bin'] = pd.cut(df['Distance_km'], bins=bins, labels=labels, include_lowest=True)
    print("Distance bins created:", labels)
else:
    print("No 'Distance_km' column found â€” skipping distance bins.")

# Quick checks / preview
print("Sample of new time features:")
display(df[['Date','year','month','day','weekday','is_weekend']].head(8))

if 'Hour' in df.columns:
    print("\nSample Hour column:")
    display(df[['Date','Hour']].head(8))

if 'distance_bin' in df.columns:
    print("\nDistribution of distance bins:")
    display(df['distance_bin'].value_counts().sort_index())
```

```python
# 2) Monthly totals (time series)
monthly_totals = df.groupby(pd.Grouper(key='Date', freq='M'))[ridership_col].sum().reset_index()
monthly_totals.columns = ['Month_End', 'Total_Passengers']
monthly_totals['Month'] = monthly_totals['Month_End'].dt.to_period('M').astype(str)
monthly_totals = monthly_totals[['Month_End','Month','Total_Passengers']]
display(monthly_totals.head(12))


# Calculate month-over-month percent change for insight
print("\nCalculating month-over-month percent change...")
monthly_totals['MoM_pct_change'] = monthly_totals['Total_Passengers'].pct_change().fillna(0) * 100
display(monthly_totals.head(12))
```

```python
# 5) Top station-pairs (From -> To)
if 'From_Station' in df.columns and 'To_Station' in df.columns:
    pair_col = df.groupby(['From_Station','To_Station'])[ridership_col].sum().reset_index()
    top_pairs = pair_col.sort_values(by=ridership_col, ascending=False).head(20)
    top_pairs.columns = ['From_Station','To_Station','Total_Passengers']
    print("\nTop 10 station pairs (From -> To):")
    display(top_pairs.head(10))
else:
    top_pairs = pd.DataFrame()
    print("\nNo station pair columns found â€” skipping station-pair aggregation.")
```
